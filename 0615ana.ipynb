```python
from sys import displayhook

import pandas as pd

# Wednesday-workingHours-stage1_flows
# firewall_test_20250731_160700-stage1_flows
df = pd.read_csv("dataset/Wednesday-workingHours-stage1_flows.csv")
df = df[(df['flow_id'] != 0) & (df['flow_start_timestamp_us'] != 0) &
        df['flow_id'].notna() & df['flow_start_timestamp_us'].notna()]
# 转换时间戳
df["datetime"] = pd.to_datetime(
    df["flow_start_timestamp_us"],
    unit="us"
)
```

```python
flow_start_time = df["datetime"].min()
flow_end_time = df["datetime"].max()

print("Flow Start:", flow_start_time)
print("Flow End:", flow_end_time)
```

```python
# 只保留恶意流
malicious = df[df["label"] == 1]
display(malicious.shape)
```

```python
start_time = malicious["datetime"].min()
end_time = malicious["datetime"].max()

print("Malicious Start:", start_time)
print("Malicious End:", end_time)
```

```python
malicious["second"] = malicious["datetime"].dt.floor("s")

result = (
    malicious
    .groupby("second")
    .size()
    .reset_index(name="malicious_flows")
)
display(result)
```

```python
df["second"] = df["datetime"].dt.floor("s")

result = (
    df.groupby(["second","label"])
      .size()
      .unstack(fill_value=0)
)

result.columns = ["benign","malicious"]
display(result)
```

```python
malicious["minute"] = malicious["datetime"].dt.floor("min")

result = (
    malicious
    .groupby("minute")
    .size()
    .reset_index(name="malicious_flows")
)

display(result)
```

```python
df["minute"] = df["datetime"].dt.floor("min")

result = (
    df.groupby("minute")["label"]
      .agg(
          total_flows="count",
          malicious_flows="sum"
      )
)

result["malicious_ratio"] = (
    result["malicious_flows"]
    / result["total_flows"]
)
display(result)
```

```python
malicious = df[df["label"] == 1].copy()

malicious = malicious.sort_values(
    "flow_start_timestamp_us"
)

malicious["delta_t"] = (
    malicious["flow_start_timestamp_us"]
      .diff()
)
display(malicious[["flow_id","delta_t"]])
```
