```python
import pandas as pd
import numpy as np

df = pd.read_csv("../dataset/ar002_et12_20260511_001-stage1_flows.csv")

# 时间排序
df = df.sort_values("flow_start_timestamp_us")

results = []

for src_ip, group in df.groupby("source_ip"):

    group = group.sort_values("flow_start_timestamp_us")

    labels = group["label"].values

    if len(labels) < 2:
        continue

    same_count = np.sum(labels[1:] == labels[:-1])

    locality_ratio = same_count / (len(labels) - 1)

    results.append({
        "source_ip": src_ip,
        "num_flows": len(labels),
        "locality_ratio": locality_ratio
    })

result_df = pd.DataFrame(results)

print(result_df["locality_ratio"].describe())

print(
    "Average Source-IP locality:",
    result_df["locality_ratio"].mean()
)
```
